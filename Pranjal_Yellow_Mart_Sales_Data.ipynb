```python
import pandas as pd
import sqlite3
import requests
from google.colab import files

# URL of the CSV file on GitHub
url = 'https://raw.githubusercontent.com/Invact-Abhay/DOE/main/Yellow%20Mart%20Sales%20Analysis.csv'

# Download the CSV file
response = requests.get(url)
with open('ylwmart.csv', 'wb') as file:
    file.write(response.content)

# Load the CSV file into a pandas DataFrame
data = pd.read_csv('ylwmart.csv')

# Create a SQLite database (or connect to an existing one)
conn = sqlite3.connect('ylwmart.db')

# Load the DataFrame into the SQLite database
data.to_sql('ylwmart', conn, if_exists='replace', index=False)
```

```python
pd.read_sql_query("select * from ylwmart",conn)
```

```python
pd.read_sql_query("""Select replace(upper(substr(FirstName,1,1))||''||lower(substr(FirstName,2))||' '||upper(substr(LastName,1,1))||''||lower(substr(LastName,2)),'_','') as fullname,
                     replace(lower(FirstName)||'.'||lower(LastName)||'@yellowmart.com','_','') as emailID from ylwmart""",conn)
```

```python
pd.read_sql_query("""Select replace(upper(substr(FirstName,1,1))||''||lower(substr(FirstName,2))||' '||upper(substr(LastName,1,1))||''||lower(substr(LastName,2)),'_','') as fullnam
                      ,Region,substr(Orderdate,-4) as Order_Year,Sales,Profit from ylwmart""",conn)
```

```python
pd.read_sql_query("""Select OrderID, length(OrderID) as len from ylwmart""",conn)
```

```python
pd.read_sql_query("""Select OrderID,length(OrderID) as len,SubCategory from ylwmart where length(OrderID)=6""",conn)
```

```python
pd.read_sql_query("""Select SubCategory,trim(Category) as Category,Sales,Profit from ylwmart where length(OrderID)=6""",conn)
```

```python
pd.read_sql_query("""Select distinct SubCategory,trim(Category) as Category from ylwmart where SubCategory like 'O%'""",conn)
```

```python
pd.read_sql_query("""Select trim(Category) as Category,sum(Cost) as TotalCost,sum(Sales) as TotalSales,sum(Profit) as TotalProfit from ylwmart group by Category""",conn)
```

```python
pd.read_sql_query("""Select substr(OrderDate,-4) as OrderYear,trim(Category) as Category,sum(Sales) as TotalSales,
                      sum(Profit) as TotalProfit from ylwmart where substr(OrderDate,-4)>=2018 group by OrderYear""",conn)
```

```python
pd.read_sql_query("""Select max(Sales) as MaxSale, substr(OrderDate,-4) as OrderYear from ylwmart group by OrderYear
                     order by OrderYear asc""",conn)
```

```python
pd.read_sql_query("""Select distinct(trim(Category)) as UniqueCategory,substr(OrderDate,-4) as OrderYear,count(OrderID) as OrderCount from ylwmart where OrderYear like '%2020%'
                      group by UniqueCategory""",conn)
```

```python
pd.read_sql_query("""Select trim(Region) as Region,substr(OrderDate,-4) as OrderYear,sum(Sales) as TotalSales from ylwmart where Region in ('East','West')
                      group by Region,OrderYear order by Region asc ,OrderYear asc""",conn)
```
